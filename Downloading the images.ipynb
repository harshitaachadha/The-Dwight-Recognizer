### __1. Downloading the .csv from google images__

__Step 1:__<br>
The firts step is to open chrome. Next open google images at this url: https://images.google.com/?gws_rd=ssl
In the search console, type in the keyword related to the images you want to download (for instance I put in Dwight Schrute) and hit enter. Once you've done that, scroll down to the bottom of the page until no more pictures automatically load and you can see the load more images button at the bottom. Now at this point, you may choose to click on the button to obtain even more urls. The main point is to scroll down through as many images as you want to download.

__Step 2:__ <br>
The next step is to navigate to the top of the screen and hit the View -> Developer -> Javascript Console options. Once you do that the javascript developer console will show up. Once that happens, copy paste the following code into the console and hit enter:

- var urls=Array.from(document.querySelectorAll('.rg_i')).map(el=> el.hasAttribute('data-src')?el.getAttribute('data-src'):el.getAttribute('data-iurl')); var hiddenElement = document.createElement('a'); hiddenElement.href = 'data:text/csv;charset=utf-8,' + encodeURI(urls.join('\n')); hiddenElement.target = '_blank'; hiddenElement.download = 'myFile.csv'; hiddenElement.click();

__Step 3:__ <br>
Once you've done this, a csv file shall automatically download. This is the file that has the url for all of the images that you wish to download. Great! you've just completed the first step to download your image datset! Now onto the next step.

```python
import requests
import cv2
import os
from imutils import paths
url_path = open('myFile.csv').read().strip().split('\n')
total = 0
if not os.path.exists('images'):
    os.mkdir('images')
image_path = 'images'
for url in url_path:
    try:
        req = requests.get(url, timeout=60)
        file_path = os.path.sep.join([image_path, '{}.jpg'.format(
            str(total).zfill(6))]
        )
        file = open(file_path, 'wb')
        file.write(req.content)
        file.close()
        print('Downloaded {}'.format(file_path))
        total += 1
    except:
        print('Could not download {}. Downloading next file'.format(file_path))
```
